```python
class HashMap:
  
    class Entry:
        def __init__(self, key, value):
            #TODO store key and value as attributes
            self.key = key
            self.value = value
             
            
        def get_key(self):
            #TODO return key
            return self.key
            
        def get_value(self):
            #TODO return value
            return self.value
            
        def __eq__(self, other):
            #TODO реализовать функцию сравнения
            #Пояснения: self - текущий объект entry, other - какой-то другой объект
            #Какое условие на равенство объектов в ассоциативном массиве?
            #Если объекты равны, то метод должен возвращать True, иначе False
            if self.key == other.key:
                return True
            return False
    
    def __init__(self, bucket_num = 64):
        #TODO Создать массив для бакетов
        self.bucket_num = bucket_num
        self.bucket_array = [None] * self.bucket_num
        self.EntryCounter = 0
        self.capacity_border = 2/3
        
    def copy_object(self):
        self.bucket_num *= 2
        new_bucket = HashMap(self.bucket_num)
        for entry in self.items():
            new_bucket.put(*entry)
        self.bucket_array = new_bucket.bucket_array
            
        
    def resize(self):
        if self.EntryCounter >= self.bucket_num*self.capacity_border:
            self.copy_object()
        

    def get(self, key, default_value=None):
        #TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        temp_obj = self.Entry(key, 0)
        index = self._get_index(self._get_hash(key))
        if self.bucket_array[index] is not None:
            for obj in self.bucket_array[index]:
                if obj == temp_obj:
                    return obj.value
        return default_value

    def put(self, key, value):
        #TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет
        temp_obj = self.Entry(key, 0)
        index = self._get_index(self._get_hash(key))
        if self.bucket_array[index] is None:
            self.bucket_array[index] = [self.Entry(key, value)]
        else:
            for obj in self.bucket_array[index]:
                if obj == temp_obj:
                    obj.value = value
                    return
            self.bucket_array[index] = [*self.bucket_array[index], self.Entry(key, value)]
        self.EntryCounter += 1
        self.resize()
        

    def __len__(self):
        #TODO Возвращает количество Entry в массиве
        return self.EntryCounter

    def _get_hash(self, key):
        #TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key)

    def _get_index(self, hash_value):
        #TODO По значению хеша вернуть индекс элемента в массиве
        return (hash_value % self.bucket_num)
        
    def values(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор значений (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        array_gen = (array_obj for array_obj in self.bucket_array if array_obj is not None)
        for array_obj in array_gen:
            for obj in array_obj:
                yield obj.value
        
    def keys(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор ключей (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        array_gen = (array_obj for array_obj in self.bucket_array if array_obj is not None)
        for array_obj in array_gen:
            for obj in array_obj:
                yield obj.key            
            
        
    def items(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор пар ключ и значение (tuples)
        #(на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        array_gen = (array_obj for array_obj in self.bucket_array if array_obj is not None)
        for array_obj in array_gen:
            for obj in array_obj:
                yield tuple([obj.key, obj.value])
        
    def __str__(self):
        #TODO Метод, вызываемый при вызове print, пофантазируйте, что можно написать
        return 'Entries: {}'.format(self.EntryCounter)
```
